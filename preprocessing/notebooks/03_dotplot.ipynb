```python
import os, glob
import numpy as np
import pandas as pd
import scanpy as sc
import matplotlib.pyplot as plt
from scipy.stats import zscore
```

```python
pth = os.path.join('..', '..')
pth_data = os.path.join(pth, 'data')
pth_in = os.path.join(pth_data, 'modeling', 'inputs')
pth_feat = os.path.join(pth_data, 'features', 'biomart')
pth_out = os.path.join(pth, 'figures', 'preprocessing', 'expression')
```

```python
adata = sc.read_h5ad(os.path.join(pth_in, 'development.h5ad'))
adata
```

```python
feat_fn = sorted(glob.glob(os.path.join(pth_feat, '*.csv')))
df = pd.read_csv(os.path.join(pth_data, 'summary.csv'), index_col = 0)
get_source = lambda fn: os.path.split(fn)[1].replace('.csv', '')
feat_dict = {get_source(fn) : fn for fn in feat_fn if 'union' not in fn}
for src, fn in feat_dict.items():
    print(src)
    df_feat = pd.read_csv(fn)
    g = df_feat.mmusculus.copy()
    g = g.loc[g.isin(adata.var_names)]
    for key, df_group in adata.obs.groupby(['source', 'celltype'], observed = True):
        msk_geo = adata.obs_names.isin(df_group.index)
        adata_geo = adata[msk_geo].copy()
        sc.tl.score_genes(adata_geo,
                          gene_list = g,
                          ctrl_as_ref = False,
                          score_name = src,
                          random_state = 1234)
        adata.obs.loc[msk_geo, src] = adata_geo.obs[src]
    # adata.obs[src] = zscore(adata.obs[src])
```

```python
feat_plt = [
    'GSE162534',
    'GSE229103',
    'GSE136689',
    'HCAEmbryonicHindlimb',
    'rRNAModifications',
    'GSE201257',
    'EMTHallmark']

msk = (adata.obs.celltype != 'nan')

df = adata[msk].obs.groupby(['source', 'celltype'], observed = True)[feat_plt].mean().T

for ix in df.index:
    df.loc[ix] = zscore(df.loc[ix])

for col in df.columns:
    df[col] = zscore(df[col])

df = df.loc[:, feat_plt[:-1]]


import seaborn as sns

fig, ax = plt.subplots(1, 1, figsize = (20, 2.5))
sns.heatmap(df, cmap = 'RdBu_r', square = True, ax = ax)

plt.show()
```
