```python
import os, glob
import numpy as np
import pandas as pd
import scanpy as sc
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import zscore
```

```python
pth = os.path.join('..', '..')
pth_data = os.path.join(pth, 'data')
pth_in = os.path.join(pth_data, 'modeling', 'inputs')
pth_feat = os.path.join(pth_data, 'features', 'biomart')
pth_out = os.path.join(pth, 'figures', 'preprocessing', 'modules')
```

```python
adata = sc.read_h5ad(os.path.join(pth_in, 'development.h5ad'))
adata
```

```python
feat_fn = sorted(glob.glob(os.path.join(pth_feat, '*.csv')))
df = pd.read_csv(os.path.join(pth_data, 'summary.csv'), index_col = 0)
get_source = lambda fn: os.path.split(fn)[1].replace('.csv', '')
feat_dict = {get_source(fn) : fn for fn in feat_fn if 'union' not in fn}
for src, fn in feat_dict.items():
    print(src)
    df_feat = pd.read_csv(fn)
    g = df_feat.mmusculus.copy()
    g = g.loc[g.isin(adata.var_names)]
    for _, df_group in adata.obs.groupby(
        ['source', 'celltype'], observed = True):
        msk = adata.obs_names.isin(df_group.index)
        adata_group = adata[msk].copy()
        sc.tl.score_genes(adata_group,
                          gene_list = g,
                          ctrl_as_ref = False,
                          score_name = src,
                          random_state = 1234)
        adata.obs.loc[msk, src] = adata_group.obs[src]
```

```python
feat_plt = [
    'GSE162534',
    'GSE229103',
    'GSE136689',
    'HCAEmbryonicHindlimb',
    'rRNAModifications',
    'GSE201257',
    'EMTHallmark']

# average signatures by cell type
msk = (adata.obs.celltype != 'nan')
df_plt = (adata[msk].obs.groupby(
          ['source', 'celltype'], observed = True
          )[feat_plt].mean().loc[feat_plt[:-1]].T)
df_plt.columns = df_plt.columns.get_level_values(1)

# normalize signatures across celltypes
df_norm = pd.concat({'mean' : df_plt.mean(axis = 1),
                     'std'  : df_plt.std(axis = 1)}, axis = 1)
df_plt = df_plt.sub(df_norm['mean'], axis = 0).div(df_norm['std'], axis = 0)
df_norm.to_csv(os.path.join(pth_data, 'figures', 'signatures_norm.csv'))

# normalize signatures across modules
for col in df_plt.columns: df_plt[col] = zscore(df_plt[col])

vlim = 2.5
fig, ax = plt.subplots(1, 1, figsize = (15, 5))
df_plt.index = df.loc[df_plt.index].Name
hm = sns.heatmap(
    df_plt,
    cmap = 'RdBu_r',
    vmin = -vlim,
    vmax = vlim,
    square = True,
    cbar = False,
    ax = ax)

ax.set_xlabel('Cell Type', size = 12)
ax.set_ylabel('Module', size = 12, labelpad = 10)
ax.tick_params(size = 0, pad = 7, labelsize = 10.5)
ax.set_xticklabels(ax.get_xticklabels(),
                   size = 10.25, rotation = 45,
                   rotation_mode = 'anchor',
                   ha = 'right', va = 'center')
fn = os.path.join(pth_out, 'signatures.pdf')
plt.savefig(fn, bbox_inches = 'tight')
```
